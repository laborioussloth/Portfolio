```python
data['Value'] = data['Value'].str[1:]
data['Wage'] = data['Wage'].str[1:]
data['Release Clause'] = data['Release Clause'].str[1:]
```

```python
def calculate(input):
    x = input.str[:-1].str.strip()
    x[x==""] = 0
    x = x.astype("float")
    y = input.str[-1]
    c = lambda x, y : x*1000 if y =="K" else x*1000000
    r = list(map(c, x, y))
    return r
data.Value = calculate(data.Value)
data.Wage = calculate(data.Wage)
data["Release Clause"] = calculate(data["Release Clause"])
```

1. Age spectrum of players, player value spectrum (Value)
2. How many players have more than 10 years of experience in the club?
3. Best player: Based on overall index (Overall) or potential (Potential).
4. Strongest Club: Calculate the average of the overall index or potential of the players in each club.
5. Find out which factors have the greatest impact on transfer value.

```python
from datetime import datetime
datetime.now().year
data.Joined = pd.to_datetime(data.Joined)
data.Tn = datetime.now().year - data.Joined.dt.year
```

```python
# List players have more than 10 years of experience in the club?
data[data.Tn >=10]
```

```python
# Age spectrum of players
plt.hist(data.Age, edgecolor = "black")
plt.xlabel('Age')
plt.ylabel('Sum')
plt.show()
```

```python
# player POT spectrum 
plt.hist(data['POT'], edgecolor = "black")
plt.xlabel('Age')
plt.ylabel('Sum')
plt.show()
```

```python
# Top 10 players have the most POT
data.nlargest(10, 'POT', keep='first').sort_values('POT', ascending = False)
```

```python
# Top 10 players have the most Average of OVA and POT
data["Ave_OVA_POT"] = (data.POT + data['OVA'])/2
data.nlargest(10, 'Ave_OVA_POT', keep='first').sort_values('Ave_OVA_POT', ascending = False)
```

```python
# Top 10 strongest club: Calculate the average of the overall index or potential of the players in each club.
data["Team"]= data.Team
data[["Team","Ave_OVA_POT"]].groupby("Team").mean().nlargest(10, "Ave_OVA_POT")
```

```python
# The greatest factor impact on transfer value
correclation_matrix = data[["Age","↓OVA","POT","Wage","Value"]].corr()
correclation_matrix
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(
    correclation_matrix, 
    annot=True,            # Hiển thị giá trị tương quan trên biểu đồ
    fmt=".2f",             # Định dạng số thập phân
    cmap="coolwarm",       # Màu sắc (có thể dùng 'viridis', 'coolwarm', 'magma', v.v.)
    cbar_kws={'label': 'Correlation Coefficient'}  # Nhãn thanh màu
)
plt.title("Correlation Matrix", fontsize=16, weight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
